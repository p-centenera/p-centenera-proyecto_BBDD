```python
#Importar librerías necesarias
import pandas as pd
import mysql.connector as conn
import matplotlib.pyplot as plt
import datetime
import math
from matplotlib.ticker import MaxNLocator
```

```python
#Conexión al servidor local de bases de datos que hayamos configurado por medio de XAMPP
try:
    mydb = conn.connect(host="localhost", database = 'clases_ceu_bc_prof',user="root", passwd="",use_pure=True)
    #Definición de una query de ejemplo
    query = "select anio, avg(valor) as promedio from euribor group by anio;"
    result_dataFrame = pd.read_sql(query,mydb) # Carga de la proyección en un Dataframe de Pandas
    print(result_dataFrame)
    mydb.close() 
except Exception as e:
    mydb.close()
```

```python
# Configuración del objeto relacionado con la figura
grafico_figura = plt.figure(figsize=(10,10))
ejes = grafico_figura.add_axes([0,0,1,1])
ejes.bar(result_dataFrame['anio'],result_dataFrame['promedio'])

#Ajustamos el eje en el que representemos los años, para evitar intervalos continuos
minimum_ele = min(result_dataFrame['anio'])
maximum_ele = max(result_dataFrame['anio'])
year_axis = range(math.floor(minimum_ele), math.ceil(maximum_ele)+1)
plt.xticks(year_axis)

plt.xlabel('Año')
plt.ylabel('Valor alcanzado')
plt.tight_layout()

# Generamos un archivo PNG con la imagen a mostrar a posteriori en el notebook
plt.savefig('static/euribor.png',bbox_inches='tight',dpi=100)
plt.show()
```
